## 1. THE TOP 5 COUNTRIES ACCORDING TO THEIR HAPPINESS RANK IN 2015

### extract the data from the csv and drop the columns which are not essential and print the top 5 countries

```python
df1=pd.read_csv("E:/2015.csv")
df11=df1.drop(["Happiness Score","Standard Error", "Region", "Family", "Dystopia Residual"], axis=1)
df111=df11.head()
df111
```

### ANALYSIS OF SOME IMPORTANT ASPECTS BY CONVERTING ABOVE TABLE TO VISUAL REPRESENTATION

```python
df1111=df111.drop(["Happiness Rank"],axis=1)
df1111.set_index("Country",inplace = True)
df1111.plot.bar(figsize=(15,7))
plt.title("Data Analysis of Top 5 Happy Countries in 2015", fontname='DejaVu Sans', fontsize=25,fontweight='bold')
plt.xlabel('Country', fontsize=20)
style.use('ggplot')
```

From the above graph we can compare the GDP,life expectancy,freedom,trust and generosity of the top 5 happy countries in 2015

# DATA OF INDIA FOR 2015-17

```python
#India 2015
dfI1=df1[df1['Country']=='India']
dfp=dfI1.drop(["Happiness Score","Standard Error", "Region", "Family","Generosity", "Dystopia Residual"], axis=1)
#India 2016
dfI2=df2[df2['Country']=='India']
dfq=dfI2.drop(["Happiness Score","Region", "Lower Confidence Interval", "Upper Confidence Interval", "Family","Generosity", "Dystopia Residual"], axis=1)
#India 2017
dfI3=df3[df3['Country']=='India']
dfr=dfI3.drop(["Happiness.Score","Whisker.high", "Whisker.low", "Family","Generosity", "Dystopia.Residual"], axis=1)
dfR = dfr.rename(columns={'Happiness.Rank': 'Happiness Rank', 'Economy..GDP.per.Capita.':'Economy (GDP per Capita)' , 'Health..Life.Expectancy.':'Health (Life Expectancy)', 'Trust..Government.Corruption.':'Trust (Government Corruption)'})
dfS=pd.concat([dfp,dfq,dfR])
dfSS=dfS.reset_index(drop=True)
dfSS['year']=['2015','2016','2017']
dfSS
```

```python
dfSS.plot(kind='line',x='year',y='Happiness Rank',color='red')
style.use('fivethirtyeight')
plt.title("Happiness rank of India in 2015-2017", fontname='DejaVu Sans', fontsize=25,fontweight='bold')
plt.xlabel('Year', fontsize=20)
plt.show()
```

```python
dfSS.plot(kind='line',x='year',y='Economy (GDP per Capita)',color='blue')
style.use('fivethirtyeight')
plt.title("GDP per capita of India in 2015-2017", fontname='DejaVu Sans', fontsize=25,fontweight='bold')
plt.xlabel('Year', fontsize=20)
plt.show()
```

increasing GDP per capita of india.

```python
dfSS.plot(kind='line',x='year',y='Health (Life Expectancy)',color='green')
style.use('fivethirtyeight')
plt.title("Life expectancy of India in 2015-2017", fontname='DejaVu Sans', fontsize=25,fontweight='bold')
plt.xlabel('Year', fontsize=20)
plt.show()
```

decrease and then slight increase of life expectancy in india.

```python
dfSS.plot(kind='line',x='year',y='Freedom',color='magenta')
style.use('fivethirtyeight')
plt.title("Sense of freedom of India in 2015-2017", fontname='DejaVu Sans', fontsize=25,fontweight='bold')
plt.xlabel('Year', fontsize=20)
plt.show()
```

increasing sense of freedom in india.

```python
dfSS.plot(kind='line',x='year',y='Trust (Government Corruption)',color='purple')
style.use('fivethirtyeight')
plt.title("Trust against Government Corruption of India in 2015-2017", fontname='DejaVu Sans', fontsize=25,fontweight='bold')
plt.xlabel('Year', fontsize=20)
plt.show()
```

increasing trust against government corruption in india.
